```python
d = 0.5
t = arange(0, d, 1/rate)
numHarmonics = 8
fundamental = 110
mySound = .5*sin(2*pi * t * fundamental)

for i in range (2, numHarmonics+1) : # for loop starting at 2 and ending at numHarmonics
    print "Adding Harmonics: ", i
    mySound = concatenate([mySound, .5*sin(2*pi * t * fundamental*i)])
    
play(mySound)
plot(mySound[0:1000]) #plot first 1000 samples
```

```python
t2 = arange(0, 1, 1/rate)

mySound2 = .5*sin(fundamental* 2*pi * t2)
for i in range(2, numHarmonics*2 + 1):
    mySound2 += .5*sin(fundamental*i * 2*pi * t2)
    

mySound2 /= numHarmonics*2 + 1
plot(mySound2[0:1000])
play(mySound2)
```

***************************************

To create a square wave with any fundamental frequency  $f$ , the  $k^{th}$  sinusoid in your sum will have frequency  $(2k-1)\times f$ . This  $k^{th}$ sinusoid will have amplitude $\frac{1}{(2k-1)}$.

In other words, the final waveform can be expressed by the following summation:

$$ s = \sum_{k=1}^{\infty} \frac{\sin(2\pi(2k-1)ft)}{2k-1}$$

```python
fun = 110
mySq = .5*sin(t2* 2*pi * fun)

k = 1000+1
for i in range(2, k) :
    mySq += .5*sin(t2* 2*pi * fun*i)
mySq /= 2*k-1


plot(mySq[0:1000])
play(mySq)
```

```python
tim = arange(0, 1, 1/100)  
x1 = .5*sin(2*2*pi*tim)
```

 FFT Math:
 
 $  A_k = \sum_{m=0}^{n-1} a_m \exp\left\{-2\pi i{mk \over n}\right\}
   \qquad k = 0,\ldots,n-1. $

```python
?fft.fft
```

```python
spect = fft.fft(x1)

print "The first spectrum element is ", spect[0]
print "The magnitude of the first element is ", abs(spect[0])
print "The length of the spectrum is ", size(spect)
```

```python
magSpect = abs(spect)
siz = int(size(magSpect)/2)

#plot(magSpect[0:siz])
stem(magSpect[0:siz])
```
